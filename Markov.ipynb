```python
import markovify
```

## Philosophy of Andrew Tate

```python
with open("andrew.txt",encoding="utf8") as f:
    text_andrew = f.read()
text_model_andrew = markovify.Text(text_andrew)

```

```python
# Generate a sentence
sentence_andrew = text_model_andrew.make_sentence()

# Generate a short paragraph
for i in range(10):
    print(text_model_andrew.make_short_sentence(max_chars=2000))
```
